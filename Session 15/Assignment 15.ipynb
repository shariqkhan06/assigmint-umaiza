#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

#### **4.Show The Top 10 highest Paid Employees**

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
import pandas as pd
import sqlite3 as sql
```

```python
con = sql.connect('database.sqlite')
con
```

```python
pd.read_sql_query('select * from salaries', con)
```

```python
#1
pd.read_sql_query("""
    SELECT JobTitle, AVG(TotalPayBenefits) AS AvgPay
    FROM salaries
    WHERE JobTitle NOT IN ('Not Provided', 'Not provided')
    GROUP BY JobTitle;
""", con)
```

```python
#2
pd.read_sql_query("""
    SELECT JobTitle, OvertimePay
    FROM salaries
    WHERE OvertimePay > 50000;
""", con)
```

```python
#3
pd.read_sql_query("""
    SELECT COUNT(DISTINCT EmployeeName) AS UniqueEmployees
    FROM salaries
    WHERE EmployeeName NOT IN ('Not Provided', 'Not provided');
""", con)
```

```python
#4
pd.read_sql_query("""
    SELECT EmployeeName, TotalPayBenefits
    FROM salaries
    ORDER BY TotalPayBenefits DESC
    LIMIT 10;
""", con)
```

```python
#5
pd.read_sql_query("""
    SELECT EmployeeName, JobTitle, TotalPayBenefits
    FROM salaries
    ORDER BY TotalPayBenefits DESC;
""", con)
```
