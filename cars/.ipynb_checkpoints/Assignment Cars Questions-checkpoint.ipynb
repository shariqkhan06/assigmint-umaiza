```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("cars.csv")
df
```

```python
#1
duplicates = df[df.duplicated()]
print("Duplicate Records:")
print(duplicates)

print("\nTotal duplicate rows:", df.duplicated().sum())
```

```python
#2
df_cleaned = df.drop_duplicates()

print("Duplicates removed:", len(df) - len(df_cleaned))
```

```python
df_cleaned.to_csv("cars_drop_duplicate.csv", index=False)

print("Cleaned dataset saved as 'cars_drop_duplicate.csv'")
```

```python
#3
print(df.isnull().sum())
```

```python
#4
df = df.dropna()
df.to_csv("cars_missing_values.csv", index=False)

print("Missing values removed & file saved as 'cars_missing_values.csv'")
```

```python
#5
top10 = df.groupby('Make')['MSRP'].mean().round().sort_values(ascending=False).head(10)

print("Top 10 Average MSRP per Car Brand:")
print(top10)
```

```python
#6
avg_msrp_per_year = df.groupby("Year")["MSRP"].mean().round(0)

print(avg_msrp_per_year)
```

```python
#7
df["Average MPG"] = (df["city mpg"] + df["highway MPG"]) / 2

result = df.groupby("Year")[["Engine HP", "Average MPG"]].mean().round(0)
print(result)
```

```python
#8
total_cars = len(df)
size_counts = df["Vehicle Size"].value_counts()
selected_sizes = size_counts.loc[["Compact", "Midsize", "Large"]]
percentages = (selected_sizes / total_cars) * 100

print(percentages.round(2))
```

```python
#9
style_counts = df["Vehicle Style"].value_counts()

print(style_counts.head())
```

```python
#10
filtered = df[(df["MSRP"] >= 40000) & (df["Popularity"] >= 1000)]
brand_counts = filtered["Make"].value_counts()

print(brand_counts.head())
```

```python
#11
df = df.dropna(subset=["Engine HP"])
plt.boxplot(df["Engine HP"])
plt.title("Boxplot of Engine HP")
plt.ylabel("Horsepower (HP)")
plt.show()
```

```python
#12
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Engine Cylinders'])
plt.title('Boxplot of Engine Cylinders')
plt.show()
```

```python
#13
df = pd.read_csv("cars.csv")
hp = df['Engine HP'].dropna()
Q1 = hp.quantile(0.25)
Q3 = hp.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = hp[(hp < lower_bound) | (hp > upper_bound)]

print(lower_bound)
print(upper_bound)
print(outliers.count())
```

```python
#14
df['Number of Doors'].fillna(df['Number of Doors'].median(), inplace=True)

print("Missing values in 'Number of Doors':", df['Number of Doors'].isnull().sum())
```
