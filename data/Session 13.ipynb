```python
(df['Region'] == 'South') & (df['Category'] == 'Furniture')
```

```python
df[(df['Region'] == 'South') & (df['Category'] == 'Furniture')]
```

```python
df[~(df['Region'] == 'South') & (df['Category'] == 'Furniture')]
```

```python
df[(df['Category'] == 'Furniture') & ~(df['Region'] == 'South')]
```
